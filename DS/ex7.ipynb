```python
from sklearn import metrics
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
model = DecisionTreeClassifier(criterion="gini")
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
clf = DecisionTreeClassifier(criterion="entropy", max_depth=4, min_samples_leaf=2)
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
report = metrics.classification_report(y_test, y_pred, target_names=iris.target_names)
print(report)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(25,20))
plot = tree.plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
```
