```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
diabetes = load_diabetes()
x = diabetes.data[:,[0,2]]
y = diabetes.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print("coefficient: ",model.coef_)
print("Intercept: ",model.intercept_)
print("mean_squared_error", mean_squared_error(y_test,y_pred))
print("R^2 score: ",r2_score(y_test,y_pred))
```

```python
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.scatter(x_test[:,0], y_test, color="red", label="Actual", alpha=0.7)
plt.scatter(x_test[:,0], y_pred, color="blue", label="Predicted", alpha=0.7)
plt.title("BMI vs Sugar level")
plt.xlabel("BMI")
plt.ylabel("Sugar level")
plt.legend()

plt.subplot(1,2,2)
plt.scatter(x_test[:,1], y_test, color="orange", label="Actual", alpha=0.7)
plt.scatter(x_test[:,1], y_pred, color="green", label="Predicted", alpha=0.7)
plt.title("Age vs Sugar level")
plt.xlabel("Age")
plt.ylabel("Sugar level")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(y_test, y_pred ,color="purple",  alpha=0.7, label="Actual vs Predicted")
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()],'r--',label="ideal fit")
plt.title("Actual vs Predicted sugar level (BMI & Age)")
plt.xlabel("Actual Sugar level")
plt.ylabel("Predicted Sugar level")
plt.legend()
plt.show()
```
