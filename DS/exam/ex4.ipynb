```python
listdata = [
    [1,"Alice",15],
    [2,"Bob",15],
    [3,"Charlie",45]
]
df = pd.DataFrame(listdata, columns=["ID","Name","Score"])
print(df)
df.to_csv("data2.csv", index=False)

sorted_df = df.sort_values(by=["Score","Name"], ascending=[False,True])
print(sorted_df)
```

```python
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Occupation": ["Engineer", "Doctor", "Engineer", "Doctor"],
    "Salary": [50000, 80000, 55000, 90000]
}
df = pd.DataFrame(data)
df.to_csv("data3.csv", index=False)
avg_salary = df.groupby("Occupation")["Salary"].mean()
print(avg_salary)
```

```python
data1 = {
    "eid":[101,102,103],
    "ename": ["Alice","Bob","Charlie"],
    "stipend": [10000, 15000, 12000]
}

data2 = {
    "eid": [101,102,103],
    "position": ["intern","Developer","trainee"]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

df_merged = pd.merge(df1, df2, on="eid")
print(df_merged)
```
