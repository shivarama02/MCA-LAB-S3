```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()
X = iris.data[:, :2]     # first two columns: sepal length, sepal width
y = iris.target          # target classes (0, 1, 2)

# Colors for each class
colors = ['red', 'green', 'blue']
labels = iris.target_names  # ['setosa', 'versicolor', 'virginica']

# Plot
plt.figure(figsize=(8,5))
for i, color in enumerate(colors):
    plt.scatter(X[y == i, 0], X[y == i, 1], 
                color=color, label=labels[i], s=50)

plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Iris Dataset - First 2 Features Scatter Plot")
plt.legend()
plt.grid(True)
plt.show()
```
