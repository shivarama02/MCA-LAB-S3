```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
men = [22,30,35,35,26]
women = [25,32,30,35,29]
group = ['G1','G2','G3','G4','G5']
```

```python
x = np.arange(len(group))
width = 0.35
```

```python
plt.figure(figsize=[10,8])
plt.bar(x-width/2, men, width, label="men", color="green")
plt.bar(x+width/2, women, width, label="women", color="red")
plt.title("Score by group and gender")
plt.xlabel("Group")
plt.ylabel("Scores")
plt.xticks(x, group)
plt.legend()
plt.show()
```

```python
data = {
    "Country":["United_States","Great_Britian","China","Russia","Germany"],
    "Gold_ Medal":[46,27,26,19,17]
}
df = pd.DataFrame(data)
df.to_csv("medal.csv", index=False)
```

```python
x = pd.read_csv("medal.csv")
counties = x["Country"]
medals = x["Gold_ Medal"]
```

```python
plt.figure(figsize=[10,8])
plt.pie(medals, labels=counties, autopct="%1.1f%%", startangle=90)
plt.title("Gold Medal Achievements - 2016 Summer Olympics")
plt.axis("equal")
plt.show()
```
