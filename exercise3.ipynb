```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,6,18])
y = np.array([3,10,12,20])
plt.plot(x,y,marker=".", color="r", mfc="g", mec="g", linestyle=":")
plt.show()
```

```python
temp = np.array([12,14,16,18,20,22,24])
sales = np.array([100,200,250,400,300,450,500])
plt.xlabel("Temperature")
plt.ylabel("Sales")
plt.plot(temp,sales,marker=".", color="b", mfc="r", mec="r", linestyle="-.")
plt.show()
```

```python
with open("array.txt","r+") as file:
    content = file.read()
    xfile,yfile = content.split("\n")
    xfile = list(map(int,xfile.split(",")))
    yfile = list(map(int,yfile.split(",")))
    plt.xlabel("Temperature")
    plt.ylabel("Sales")
    plt.plot(xfile,yfile,marker=".",ms=10, color="r", mfc="r", mec="black", linestyle="--")
    plt.show()
```
